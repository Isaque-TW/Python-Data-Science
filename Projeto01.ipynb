```python
# Importação das Bibliotecas

import pandas as pd  # Importando Biblioteca / Ferramenta para ter acesso a base de dados
import plotly.express as px # Biblioteca para construção de gráficos e Dashboards no Python
import smtplib # Biblioteca para envio de email
import email.message as em
```

```python
# PASSO 1 - Importar a base de dados (Pasta Vendas)

tabela_vendas = pd.read_excel("/content/drive/MyDrive/Python para Ciência de Dados/Vendas/Vendas+-+Base+de+Dados.xlsx")
display(tabela_vendas)
```

```python
# PASSO 2 - Calcular o produto mais vendido (em quantidade)

tb_quantidade_produto = tabela_vendas.groupby('Produto').sum()
tb_quantidade_produto = tb_quantidade_produto[['Quantidade']]
tb_quantidade_produto = tb_quantidade_produto.sort_values(by='Quantidade', ascending=False)
display(tb_quantidade_produto)
```

```python
# PASSO 3 - Calcular o produto mais vendido (em faturamento)

tabela_vendas['Faturamento'] = tabela_vendas['Quantidade'] * tabela_vendas['Valor Unitário']
tb_faturamento_produto = tabela_vendas.groupby('Produto').sum()
tb_faturamento_produto = tb_faturamento_produto[['Faturamento']].sort_values(by='Faturamento', ascending=False)
display(tb_faturamento_produto)
```

```python
# PASSO 4 - Calcular a loja/estado que mais vendeu (em faturamento)

tb_faturamento_loja = tabela_vendas.groupby('Loja').sum()
tb_faturamento_loja = tb_faturamento_loja[['Faturamento']].sort_values(by='Faturamento', ascending=False)
display(tb_faturamento_loja)
```

```python
# PASSO 5 - Calcular o ticket médio por loja/estado.

tabela_vendas['Ticket Médio'] = tabela_vendas['Valor Unitário'] # Implementa uma nova coluna (Ticket Médio) 'virtual' espelhada em ['Valor Unitário']
tb_ticket_medio = tabela_vendas.groupby('Loja').mean(numeric_only=True) #Método mean, Cria uma nova variável que recebe tabela_vendas agrupado por coluna 'Loja' e aplica a media em  todas colunas
tb_ticket_medio = tb_ticket_medio[['Ticket Médio']] # Aplicando filtro pra trazer somente o Ticket Médio
tb_ticket_medio = tb_ticket_medio.sort_values(by='Ticket Médio', ascending=False)
display(tb_ticket_medio) # Ordenando em ascending=False
```

```python
# PASSO 6 - Criar um gráfico/dashboard da loja/estado que mais vendeu (em faturamento)

# O grafico é composto por 3 argumentos: Eixo Y representado pela barra vertical; Eixo X representado pela barra horizontal para informal qual é a tabela e as lojas/estados. 1º saber a base de dados/tabelas que vai utilizar. 2º Qual eixo definir, o eixo Y é o faturamento. E para o eixo X passar as informações das lojas/estados.

grafico = px.bar(tb_faturamento_loja, y='Faturamento', x=tb_faturamento_loja.index) # Passando os 3 argumentos. 1º eixo Y tabela de faturamento (titulo da coluna). 2º A biblioteca panda interpreta as lojas/estados como index/índice como um Id e por isso se torna negrito e desalinha a coluna loja. Por isso passamos o nome da tabela no eixo X com index
grafico.show()
```

# O que são Funções?
Na programação, funções são blocos de código que realizam determinadas tarefas que normalmente precisam ser executadas diversas vezes dentro de uma aplicação. Quando surge essa necessidade, para que várias instruções não precisem ser repetidas, elas são agrupadas entuma função, à qual é dado um nome e que poderá ser chamada/executada em diferentes partes do programa.

# O que é Lambda?



*  **Função** "**Anônima**" - Não precisa definir um nome e são diretas.



      Ex: Suponhamos que você esteja avaliando o preço de um produto e quer saber quanto de imposto será cobrado sobre tal
      produto. O imposto é correspondente a 20% do valor do produto.




  ### **Função Convencional (Padrão)**





```python
preco = 500

# def=Função que pega o valor(preço) entra na função e multiplicado por 20%
def calcular_imposto(preco):
  return preco * 0.2

print(calcular_imposto(preco))
```

### **Função Lambda**

```python
preco2 = 1000
calcular_imposto2 = lambda preco2: preco2 * 0.2

print(calcular_imposto2(preco2))
```

### **Qual usar? Função Lambda ou Função Convencional?**

```python
precos = [1000, 500, 70, 10]
# O map é uma função que aceita funções dentro do seu argumento. Ele acessa cada elemento da lista - A função map pega uma lista e a transforma numa nova lista, executando algum tipo de operação em cada elemento.
impostos = list(map(lambda x: x * 0.2, precos))
# Usando Lambda para calcular porcentagem de impostos: Função lambda que recebe um valor, ou seja, o preço caracterizado como(x) que começa valendo como primeiro item da lista de impostos, retorna esse valor multiplicado pela %. O map vai fazer o calculo para cada valor/elemento da lista.
print(impostos)
```

### Qual é a vantagem do Lambda?
### R: A vandatgem do Lambda é não ter que criar primeiro uma função pra depois utiliza-la. Com Lambda passamos a função direto.


# **FORMATAÇÕES NUMÉRICAS E MONETÁRIAS**

**Loja/Estado que mais vendeu (em faturamento)**

```python
# EXEMPLO
from babel.numbers import format_currency
# Importando uma ferraenta para formatação de numeros e valores monetários da biblioteca babel.numbers

valor = 100000
valor_formatado = format_currency(valor, 'BRL', locale='pt_BR')

print(valor_formatado)
```

```python
from babel.numbers import format_currency

# Formatação monetária
tb_faturamento_loja_formatado = pd.DataFrame(tb_faturamento_loja['Faturamento'].apply(lambda x: format_currency(x, 'BRL', locale='pt_BR')))

# Método para voltar/resetar os indices/index/id(Loja) no formato numerico onde que alinha os titulos e colunas.
tb_faturamento_loja_formatado = tb_faturamento_loja_formatado.reset_index()

# Formatação HTML - Preparação HTML retirando o index com False e alinhando as palavras com justify='center' =0
tb_faturamento_loja_formatado_html = tb_faturamento_loja_formatado.to_html(index=False, justify='center', border=0).replace('<tbody>', '<tbody style="text-align: center; color: #484848; background: #F7F7F7">')

display(tb_faturamento_loja_formatado)
display(tb_faturamento_loja_formatado_html)
```

**Produtos mais vendidos (em quantidade)**

```python
qtd = 15000000

# Função format Python
# Formatação de milhares. Por default, ele segue padrão USA, com . parada casa decimais e , para milhares. - Usamos .replace para adc lingua BR - substitui o primeiro(',') pelo segundo. ) - Para adc casa decimais como por exemplo para valores monetários, usamos :    "{:,.2f}" Significa que eu qeuro separar a casa de MILHAR por vírgula e as casas DECIMAIS com . 2f (duas casas decimais) e ele é um f de float = PONTO FLUTUANTE que é o valor das casas decimais. EX: - qtd_formatado = "{:,.2f}".format(qtd)
qtd_formatado = "{:,}".format(qtd).replace(',', '.')
print(qtd_formatado)
```

```python
# Formatação numérica  - Cria a variável que recebe a tabela quantidade produtos na coluna 'Quantidade' e apply usando Lamda + valor(x) que são as Quantidades de cada produto na tabela/coluna 'Quantidade', formata para , para milhares e depois usa .replace para modificar , por ponto .
tb_quantidade_produto_formatado = pd.DataFrame(tb_quantidade_produto['Quantidade'].apply(lambda x: "{:,}".format(x).replace(',', '.')))
tb_quantidade_produto_formatado = tb_quantidade_produto_formatado.reset_index()

tb_quantidade_produto_formatado_html = tb_quantidade_produto_formatado.to_html(index=False, justify='center', border=0).replace('<tbody>', '<tbody style="text-align: center; color: #484848; background: #F7F7F7">')

display(tb_quantidade_produto_formatado)
display(tb_quantidade_produto_formatado_html)
```
